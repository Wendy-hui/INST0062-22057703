```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('result_metrics.csv',index_col = 0)
```

```python
order = ['bleu_1gram','bleu_2gram','bleu_3gram','bleu_4gram',
             'METEOR',
             'CIDEr',
             'Spice',
             'ROUGE_L','ROUGE_L_pre','ROUGE_L_recall',
             'ROUGE_1gram','ROUGE_1gram_pre','ROUGE_1gram_recall',
             'ROUGE_2gram','ROUGE_2gram_pre','ROUGE_2gram_recall',
             ]
df = df.reindex(columns=order)
df
```

```python
corr = df.corr()
corr.to_csv('corr.csv')
corr
```

```python
sns.heatmap(corr , annot=False, cmap='YlOrBr', xticklabels=1, yticklabels=1)
plt.savefig('heatmap.svg', dpi=600, format='svg', bbox_inches="tight")
plt.show()
```

```python
sns.pairplot(df)
plt.savefig('scattar.svg', dpi=600, format='svg', bbox_inches="tight")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].scatter(df['bleu_4gram'], df['METEOR'])
axes[0].scatter(x=0.42266003, y=0.306639661, c='red')
axes[0].set_xlabel("BLEU-4")
axes[0].set_ylabel("METEOR")
axes[1].scatter(df['ROUGE_L'], df['METEOR'])
axes[1].scatter(x=0.644203357, y=0.306639661, c='red')
axes[1].set_xlabel("ROUGE_L")
axes[2].scatter(df['CIDEr'], df['METEOR'])
axes[2].scatter(x=1.456246735, y=0.306639661, c='red')
axes[2].set_xlabel("CIDEr")
plt.savefig('outlier1.svg', dpi=600, format='svg', bbox_inches="tight")
plt.show()
```
